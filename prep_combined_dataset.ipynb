```python
import pandas as pd
import numpy as np
```

```python
# Define the file path (make sure to replace this with your actual file path)
file_path = 'income_statement_alibaba.csv'

# Load the CSV file into a DataFrame
alibaba_income = pd.read_csv(file_path)
# Define the file path (make sure to replace this with your actual file path)
file_path = 'income_statement_amazon.csv'

# Load the CSV file into a DataFrame
amazon_income = pd.read_csv(file_path)

# Define the file path (make sure to replace this with your actual file path)
file_path = 'income_statement_ebay.csv'

# Load the CSV file into a DataFrame
ebay_income = pd.read_csv(file_path)

# Define the file path (make sure to replace this with your actual file path)
file_path = 'income_statement_shopify.csv'

# Load the CSV file into a DataFrame
shopify_income = pd.read_csv(file_path)
```

```python
#check the dfs, found out that the currewncy of alibaba is not in USD so I will have to find the exchange rate per year
alibaba_income.head()
```

```python
# Define the file path (make sure to replace this with your actual file path)
file_path = 'CNY_to_USD.csv'

# Load the CSV file into a DataFrame
CNY_to_USD = pd.read_csv(file_path)
```

```python
CNY_to_USD.head(2)
```

```python

# Step 1: Convert the column names to datetime
CNY_to_USD.columns = pd.to_datetime(CNY_to_USD.columns, format='%Y-%m-%d')

# Step 2: Resample the columns by year and calculate the mean
CNY_to_USD_yearly = CNY_to_USD.resample('Y', axis=1).mean()

# Step 3: Rename the columns to show only the year (not full date)
CNY_to_USD_yearly.columns = CNY_to_USD_yearly.columns.year

print(CNY_to_USD_yearly)
```

```python
# Calculate the average for each year
CNY_to_USD_yearly_avg = CNY_to_USD_yearly.mean(axis=0)

# Convert to DataFrame for better readability (optional)
CNY_to_USD_yearly_avg = CNY_to_USD_yearly_avg.reset_index()
CNY_to_USD_yearly_avg.columns = ['Year', 'Average']

print(CNY_to_USD_yearly_avg)
```

```python
alibaba_income.columns
```

```python
# we need to exchange the rates from CNY to USD 
# Convert 'Fiscal Date Ending' to datetime
alibaba_income['Fiscal Date Ending'] = pd.to_datetime(alibaba_income['Fiscal Date Ending'])

# Extract the year from 'Fiscal Date Ending'
alibaba_income['Year'] = alibaba_income['Fiscal Date Ending'].dt.year

# Merge financial data with exchange rates on year
merged_data_alibaba = pd.merge(alibaba_income, CNY_to_USD_yearly_avg, on='Year', how='left')

# Convert the financial data from CNY to USD
for column in ['Gross Profit',
       'Total Revenue', 'Cost of Revenue', 'Cost of Goods and Services Sold',
       'Operating Income', 'Selling, General and Administrative',
       'Research and Development', 'Operating Expenses',
       'Investment Income Net', 'Net Interest Income', 'Interest Income',
       'Interest Expense', 'Non-Interest Income', 'Other Non-Operating Income',
       'Depreciation', 'Depreciation and Amortization', 'Income Before Tax',
       'Income Tax Expense', 'Interest and Debt Expense',
       'Net Income from Continuing Operations',
       'Comprehensive Income Net of Tax', 'EBIT', 'EBITDA', 'Net Income']:  # Add other columns as needed
    merged_data_alibaba[column + ' (USD)'] = merged_data_alibaba[column] * merged_data_alibaba['Average']

# Drop the unnecessary columns
merged_data_alibaba.drop(columns=['Average'], inplace=True)

print(merged_data_alibaba)
```

```python
merged_data_alibaba.drop(columns=['Fiscal Date Ending'], inplace=True)
```

```python
# Add company name to each DataFrame
merged_data_alibaba['Company'] = 'alibaba'
amazon_income['Company'] = 'amazon'
ebay_income['Company'] = 'ebay'
shopify_income['Company'] = 'shopify'
```

```python
   
def keep_year_remove_date_and_currency(df):
    df['Fiscal Date Ending'] = pd.to_datetime(df['Fiscal Date Ending'])

    # Extract the year from 'Fiscal Date Ending'
    df['Year'] = df['Fiscal Date Ending'].dt.year
    
    # Drop the original 'Fiscal Date Ending' column
    df.drop(columns=['Fiscal Date Ending'], inplace=True)
    # Drop the original 'Fiscal Date Ending' column
    df.drop(columns=['Reported Currency'], inplace=True)
    return df
```

```python
keep_year_remove_date_and_currency(amazon_income)
keep_year_remove_date_and_currency(ebay_income)
keep_year_remove_date_and_currency(shopify_income)
```

```python
# Remove "(USD)" from column names
merged_data_alibaba.columns = merged_data_alibaba.columns.str.replace(r' \(USD\)$', '', regex=True)
```

```python
# Combine all DataFrames into one amazon+alibaba+shopify+ebay
income_AASE = pd.concat([amazon_income, shopify_income, ebay_income, merged_data_alibaba], ignore_index=True)

# Optionally, reset index
income_AASE = income_AASE.reset_index(drop=True)
income_AASE.columns
```

```python
# Save the combined DataFrame to an Excel file
file_path = 'income_AASE.xlsx'
income_AASE.to_excel(file_path, index=False, engine='openpyxl')

print(f'Data successfully saved to {file_path}')
```
